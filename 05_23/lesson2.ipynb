for 迴圈 -> 知道要執行幾次

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,5):
    print(i)
```

```python
for i in range(1,10,3):
    print(i)
```

```python
total = 0

for i in range(50,101):
    total += i

print(total)
```

if - elif - else

```python
score = int(input("請輸入數字:"))

if score >= 90:
    print("A")

elif score >= 80:
    print("B")

elif score >= 70:
    print("C")

elif score >= 60:
    print("D")

else:
    print("F")
```
