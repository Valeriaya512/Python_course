```python
# 學生總分為300
# 有些學生可以加分5%

while True:
    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        break
    except ValueError:
        print("輸入錯誤，請輸入整數!")
is_add = input("學生是否符合加分條件?(y,n)")

if is_add == "y":   #單項選擇
    scores = scores * 1.05

# 四捨五入(5進位)
scores = int(scores + 0.5)

#超過300分，上限取300分
if scores > 300:
    scores = 300

print("學生分數:", scores)
```

```python
while True:
    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        break   # 跳出while迴圈
    except ValueError:
        print("輸入錯誤，請輸入整數!")
```

```python
i = 1

while i <= 5:  # if i>5 -> false，跳出迴圈
    print(i)
    i += 1
```

while - continue

```python
i = 1

while i <= 10:
    i += 1
    if i % 2 == 0:
        continue  # 跳回while
    print(i)
```

continue - break

```python
i =  0

while i <= 10:
    i += 1

    if i == 3:
        continue

    if i  == 8:
        break

    print(i)
```

```python
while True:
    text = input("請輸入數字，輸入q結束:")

    if text == 'q':
        break

    if text == "":
        continue

    print(text)
```
