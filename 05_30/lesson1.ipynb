```python
# 資料結構
# 程式結構

# list
# 可以修改
scores = [67, 51, 79, 81, 96]
print(scores)
print(scores[0])
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])
#print(scores[5])
scores[0] = 100
print(scores)
```

```python
# tuple
# tuple建立後，不可修改內容，read olny
# 何時使用,暫時儲存
scores1 = (67, 51, 79, 81, 96)
print(scores1)
print(scores1[0])
print(scores1[1])
print(scores1[2])
print(scores1[3])
print(scores1[4])
#scores1[0] = 100
#print(scores1)
```

```python
# dictionary
students = {'chinese':89,
    'english':94,
    'math':65,
	'history':92,
	'discover': 95}

print(students)
print(students['chinese'])
print(students['history'])
print(students['math'])
print(students['english'])
print(students['discover'])
```

資料結構

```python
# list
scores = [67, 51, 79, 81, 96]

for s in scores:
    print(s)
```

```python
# dict
students = {'chinese':89,
    'english':94,
    'math':65,
	'history':92,
	'discover': 95}

for key in students:
    #print(key)
    print(students[key])
```
