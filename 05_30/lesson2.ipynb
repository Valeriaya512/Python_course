```python
# 讀取檔案
try:
	with open("student.txt", "r", encoding="utf-8") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open(file="student.txt") 引數名稱的呼叫
try:
	with open(file="student.txt", encoding='utf-8') as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open("student.txt") 引數值的呼叫,必需依順序

try:
	with open("student.txt", 'r', -1, 'utf-8') as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open(file="student.txt") 引數名稱的呼叫,可以不依順序,只要依照名

try:
	with open(file="student.txt", encoding="utf-8") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open(file="student.txt") 混合的呼叫,可以不依順序,只要依照名

try:
	with open("student.txt", encoding="utf-8", mode="r") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
file = open("student.txt","r", encoding="utf-8")
print(type(file))
content = file.read()
print(content)

file.close()  #關閉檔案連接
print(file.closed)  #確認是否關閉
```

```python
# with 跳出程式區塊自動關閉檔案連接
with open("student.txt","r", encoding="utf-8") as file:
    content = file.read()

print(file.closed)
```
