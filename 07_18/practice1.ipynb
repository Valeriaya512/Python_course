```python
import numpy as np
from numpy import ndarray
import pandas as pd
from pandas import DataFrame

#scores:ndarray | None = None # 學生的分數
#scores_df:DataFrame | None = None # 學生分數的DataFrame

scores:ndarray = np.random.randint(40,101, size=(50, 5))
scores_df:DataFrame = pd.DataFrame(scores,            
            columns=['國文', '英文', '數學', '地理', '歷史'],
            index=range(1,51))


#scores_df.T  #轉置

names_df:DataFrame = pd.read_csv('students.csv')
scores_df[['姓名','性別']] = names_df[['姓名','性別']].head(n=50).values   #第一層[]是subscript,第二層[]才是list
scores_df = scores_df[['姓名', '性別','國文', '英文', '數學', '地理', '歷史']]   #變換欄位
#scores_df

#選取欄
scores_df[['姓名','國文']]
#選取列
scores_df.iloc[5]
scores_df.iloc[5:10]  #silce
```

```python
# 透過值來搜尋
scores_df[scores_df['姓名'] == '許美慧']
scores_df[scores_df['國文'] < 60]
scores_df.query("性別 == '女' and 國文 < 60")
```

```python
# 統計
scores_df["總分"] = scores_df[['國文', '英文', '數學', '地理', '歷史']].sum(axis=1)  # axis=1：橫向加總
scores_df["平均"] = scores_df[['國文', '英文', '數學', '地理', '歷史']].mean(axis=1)
scores_df["排名"] = scores_df['總分'].rank(ascending = False, method = 'min')
scores_df.sort_values(by="排名")
```

```python
from pandas import Series
def calculate(s:Series) -> int:
    #print(s)
    #print((~(s < 60)).sum())  #計算不及格的科目數
    n = (s < 60).sum()    #計算及格的科目數
    m = (~(s < 60)).sum()
    return n,m

#scores_df['及格的科目數'] = scores_df[['國文', '英文', '數學', '地理', '歷史']].apply(calulate,axis=1)
#scores_df['不及格的科目數'] = scores_df[['國文', '英文', '數學', '地理', '歷史']].apply(calulate,axis=1)
#scores_df
#修改
result = scores_df[['國文', '英文', '數學', '地理', '歷史']].apply(calculate, axis=1, result_type='expand')
result.columns = ['及格的科目數', '不及格的科目數']
scores_df = pd.concat([scores_df, result], axis=1)
scores_df
```

```python
def get_failing_subjects(row):
    subjects = ['國文', '英文', '數學', '地理', '歷史']
    failing = [subj for subj in subjects if row[subj] < 60]    
    return ','.join(failing) if failing else "無"
    
    
scores_df["不及格的科目名稱"] = scores_df[['國文', '英文', '數學', '地理', '歷史']].apply(get_failing_subjects,axis=1)
scores_df
```
